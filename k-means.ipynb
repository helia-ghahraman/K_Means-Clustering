```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics.cluster import rand_score
```

```python
iris = datasets.load_iris(as_frame=True)
x = iris.data
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_init=10, n_clusters=i)
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 10), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans = KMeans(n_init=10, n_clusters=3)
clusters = kmeans.fit_predict(x_train)
clusters = clusters.tolist()
y_train_list = y_train.tolist()
rand_score(clusters, y_train_list)
```

```python
kmeans = KMeans(n_init=10, n_clusters=3)
clusters = kmeans.fit_predict(x_test)
clusters = clusters.tolist()
y_test_list = y_test.tolist()
rand_score(clusters, y_test_list)
```

```python
kmeans = KMeans(n_init=10, n_clusters=3)
clusters = kmeans.fit_predict(x)
clusters = clusters.tolist()
y = y.tolist()
rand_score(clusters, y)
```

```python
plt.scatter(x['petal length (cm)'], x['petal width (cm)'], c=clusters, cmap='rainbow')
```
